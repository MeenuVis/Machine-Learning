```python
#Pie plot for attrition
fig = px.pie(df,names='Attrition',title='Attrition',color_discrete_sequence=px.colors.sequential.Aggrnyl)
fig.show()
```

```python
#Plot for attrition and age
print(df[(df['Attrition'] == 'Yes')].groupby('Age')['Age'].count().sort_values(ascending=False))
plt.figure(figsize=(14,6), dpi=80)
sns.countplot(data=df, x='Age', hue='Attrition', order = df['Age'].value_counts().index, palette='seismic_r').set_title
('Attrition by Age');
```

```python
#Plot for distance and attrition
plt.subplot(2,1,2)
sns.countplot(x= 'DistanceFromHome' ,data = df ,palette='winter_r'  ,hue =df['Attrition'])
```

```python
#Plot for marital status and attritiom
sns.countplot(data=df, x="MaritalStatus", hue="Attrition")
```

```python
#Plot for identifying all departments
fig = px.pie(df, names='Department', title='Department',color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

```python
#Ploting department and attrition
sns.countplot(data=df, x="Department", hue="Attrition")
```

```python
#plot for job role and attrition
sns.countplot(y= 'JobRole' ,data = df ,palette='winter_r'  ,hue ='Attrition')
```

```python
#Ploting over tie and attrition
sns.countplot(data=df, x="OverTime", hue="Attrition")
```

```python
#Business travel and attrition plot
b=sns.countplot(x= 'BusinessTravel' ,data = df ,palette='hls'  ,hue ='Attrition')
for p in b.patches:
    x = p.get_x() + p.get_width()
    y = p.get_height()
plt.show()
```

```python
#Conveting Variable to categorical value for better ploting
df['Education'].replace([1,2,3,4,5],["Below College","College"," Bachelor","Master","Doctor"],inplace=True)
df['EnvironmentSatisfaction'].replace([1,2,3,4,],["Low ","Medium"," High","Very High"],inplace=True)
df['JobInvolvement'].replace([1,2,3,4],["Low ","Medium"," High","Very High"],inplace=True)
df['JobSatisfaction'].replace([1,2,3,4],["Low ","Medium"," High","Very High"],inplace=True)
df['PerformanceRating'].replace([1,2,3,4],["Low ","Good"," Excellent"," Outstanding"],inplace=True)
df['RelationshipSatisfaction'].replace([1,2,3,4],["Low ","Medium"," High","Very High"],inplace=True)
df['WorkLifeBalance'].replace([1,2,3,4,],["Bad ","Good"," Better"," Best"],inplace=True)
df.head()
```

```python
#Education and Attrition plot
sns.countplot(x= 'Education' ,data =df ,palette='pastel',hue='Attrition')
plt.title('Education Vs Attrition')
plt.show()
```

```python
#Educational field and attrition
plt.figure(figsize=(9,5))
sns.countplot(x= 'EducationField' ,data =df ,palette='winter_r',hue='Attrition')
plt.title('EducationField Vs Attrition')
plt.show()
```

```python
#Plot for attrition and year with manager
sns.kdeplot(data=df, x="YearsWithCurrManager", hue="Attrition")
```
